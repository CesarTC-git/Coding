```python
from selenium import webdriver
from selenium.webdriver.support.ui import WebDriverWait, Select
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
import time
import pandas as pd
import requests

options = webdriver.ChromeOptions()
options.add_argument('--start-maximized')
options.add_argument('--disable-extensions')

driver_path = 'C:\\Users\\edal_\\Downloads\\chromedriver.exe'

# GET INTO THE WEBPAGE
driver = webdriver.Chrome(driver_path, options=options) # set up driver
#driver.set_window_position(2000,0)
driver.maximize_window()
time.sleep(1)

driver.get('https://xeno-canto.org/explore?query=cnt%3A%22%3DSpain%22+seen%3A%22yes%22+grp%3A%22birds%22&pg=1') # open website

#driver.get('https://xeno-canto.org/explore?query=box%3A35.631%2C-10.626%2C43.858%2C3.656+&pg=19') # starting on page 19

domain = 'https://xeno-canto.org/'

count = 0
df = pd.DataFrame()
for j in range(150):
    for i in range(1,30,1):
        try:
            name = driver.find_element(By.XPATH, '//*[@id="content-area"]/table/tbody/tr[{}]/td[2]/span[1]'.format(i)).text
            scientific_name = driver.find_element(By.XPATH, '//*[@id="content-area"]/table/tbody/tr[{}]/td[2]/span[2]'.format(i)).text
            duration    = driver.find_element(By.XPATH,'//*[@id="content-area"]/table/tbody/tr[{}]/td[3]'.format(i)).text
            recorded_by = driver.find_element(By.XPATH,'//*[@id="content-area"]/table/tbody/tr[{}]/td[4]'.format(i)).text
            date = driver.find_element(By.XPATH,'//*[@id="content-area"]/table/tbody/tr[{}]/td[5]'.format(i)).text
            country = driver.find_element(By.XPATH,'//*[@id="content-area"]/table/tbody/tr[{}]/td[7]'.format(i)).text
            area = driver.find_element(By.XPATH,'//*[@id="content-area"]/table/tbody/tr[{}]/td[8]'.format(i)).text
            altitude = driver.find_element(By.XPATH,'//*[@id="content-area"]/table/tbody/tr[{}]/td[9]'.format(i)).text
            type_of_sound = driver.find_element(By.XPATH,'//*[@id="content-area"]/table/tbody/tr[{}]/td[10]'.format(i)).text
            href = driver.find_element(By.XPATH,'//*[@id="content-area"]/table/tbody/tr[{}]/td[12]/a[1]'.format(i))
            key = driver.find_element(By.XPATH,'//*[@id="content-area"]/table/tbody/tr[{}]/td[13]/a[1]'.format(i)).text

            new_row = {'Code':key,'Name':name, 'Scientific_name':scientific_name,'Duration':duration, 'Recorded_by':recorded_by, 'Date':date, 'Country':country, 'Area':area,'Altitude':altitude,'Type_of_sound':type_of_sound}
            df = df.append(new_row, ignore_index=True)

            with open(key + '.mp3', "wb") as file:
                response = requests.get(domain + key[2:] + '/download')
                file.write(response.content)
        except:
            continue
    count += 1
    if j == 0:
        element_SIGUIENTE = '//*[@id="content-area"]/nav[1]/ul/li[12]/a'
    elif j<5:
        element_SIGUIENTE = '//*[@id="content-area"]/nav[1]/ul/li[13]/a'
    elif j == 5:
        element_SIGUIENTE = '//*[@id="content-area"]/nav[1]/ul/li[14]/a'
    else:
        element_SIGUIENTE = '//*[@id="content-area"]/nav[1]/ul/li[15]/a'
        
    WebDriverWait(driver,5)\
        .until(EC.element_to_be_clickable((By.XPATH, element_SIGUIENTE)))\
        .click() 
    
driver.quit()

df.to_csv('Birds.csv')
```
